### Seleção de frequências

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')
```

```python
dados.head(10)
```

```python
#Selecione somente os imóveis classificados com tipo 'Apartamento'.
selecao = dados['Tipo'] == 'Apartamento'
n1 = dados[selecao].shape[0]
n1
```

```python
#Selecione os imóveis classificados com tipos 'Casa', 'Casa de Condomínio' e 'Casa de Vila'.
selecao = (dados['Tipo'] == 'Casa') | (dados['Tipo'] == 'Casa de Condomínio') | (dados['Tipo'] == 'Casa de Vila')
n2 = dados[selecao].shape[0]
n2 
```

```python
#Selecione os imóveis com área entre 60 e 100 metros quadrados, incluindo os limites.
# 60 < = Area <= 100
selecao = (dados['Area'] >= 60)&(dados['Area'] <= 100)
n3 = dados[selecao].shape[0]
n3 
```

```python
#Selecione os imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00.
selecao = (dados['Quartos'] >= 4) & (dados['Valor'] < 2000)
n4 = dados[selecao].shape[0]
n4 
```

```python
print(f"Nº de imóveis classificados com tipo 'Apartamento' -> {n1}")
print(f"Nº de imóveis classificados com tipos 'Casa', 'Casa de Condomínio' e 'Casa de Vila'-> {n2}")
print(f"Nº de imóveis com área entre 60 e 100 metros quadrados, incluindo os limites -> {n3}")
print(f"Nº de imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00 -> {n4}")
```

# Aula 9

* Nesta aula, aprendemos:

* Como identificar e remover outliers com o box plot
* Como fazer um gráfico de pizza com a aplicação do método pie(), da biblioteca matplotlib


```python
plt.rc('figure', figsize = (14, 6))
dados_geral = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')
```

```python
dados_geral.boxplot(['Valor'])
```

```python
dados_geral[dados_geral['Valor'] >= 500000]
```

```python
valor = dados_geral['Valor']
```

```python
Q1 = valor.quantile(.25)
Q1
```

```python
Q3 = valor.quantile(.75)
Q3
```

```python
IIQ = Q3 - Q1
IIQ
```

```python
limite_inferior = Q1 - 1.5 * IIQ
limite_superior = Q3 + 1.5 * IIQ
```

```python
selecao = (valor >=limite_inferior) & (valor <= limite_superior)
dados_new = dados_geral[selecao]
```

```python
dados_new.boxplot(['Valor'])
```

```python
dados_geral.hist(['Valor'])
dados_new.hist(['Valor'])
```

```python
dados_geral.boxplot(['Valor'], by = ['Tipo'])
```

```python
grupo_tipo = dados_geral.groupby(['Tipo'])
```

```python
type(grupo_tipo)
```

```python
grupo_tipo = dados_geral.groupby('Tipo')['Valor']
```

```python
type(grupo_tipo)
```

```python
grupo_tipo.groups
```

```python
Q1 =  grupo_tipo.quantile(.25)
Q3 =  grupo_tipo.quantile(.75)
IIQ = Q3 - Q1
limite_inferior = Q1 - 1.5 * IIQ
limite_superior = Q3 + 1.5 * IIQ
```

```python
Q1
```

```python
Q3
```

```python
IIQ
```

```python
limite_inferior
```

```python
limite_superior
```

```python
limite_superior['Apartamento']
```

```python
limite_superior['Casa']
```

```python
for tipo in grupo_tipo.groups.keys():
    print(tipo)
```

```python
dados_new = pd.DataFrame()
for tipo in grupo_tipo.groups.keys():
    eh_tipo = dados_geral['Tipo'] == tipo
    eh_dentro_limite = (dados_geral['Valor'] >= limite_inferior[tipo]) & (dados_geral['Valor'] <= limite_superior[tipo])
    selecao = eh_tipo & eh_dentro_limite
    dados_selecao = dados_geral[selecao]
    dados_new = pd.concat([dados_new, dados_selecao])
```

```python
dados_new.boxplot(['Valor'], by = ['Tipo'])
```

```python
dados_new.to_csv('dados/aluguel_residencial_sem_outliers.csv', sep = ';', index = False)
```
